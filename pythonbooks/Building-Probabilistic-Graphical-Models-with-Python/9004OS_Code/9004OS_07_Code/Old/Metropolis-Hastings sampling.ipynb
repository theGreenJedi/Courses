In this code snippet, we'll run the Metropolis-Hastings algorithm.

```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from math import *
from random import random

def unif(a,b):
  return a+(b-a)*random()

def sdnorm(z):
  return exp(-z*z/2)

n=900
alpha=1
x=0
candarr=[]

for i in range(n):
    innov=unif(-alpha,alpha)
    #new proposal 
    can=x+innov
    #acceptance probability
    aprob=min(1,sdnorm(can)/sdnorm(x))
    u=random()
    #accept the proposal is a random coin flip is less than the acceptance probability
    if (u<aprob):
        #move to the new location.
        x=can
        candarr.append((can,'g'))
    else:
        candarr.append((can,'r'))

#plot the accepted and rejected candidates, and a histogram of the accepted candidates.
f, axarr = plt.subplots(2, 1)
f.set_size_inches(10,8)
col=[i[1] for i in candarr]
axarr[0].scatter(range(n),[i[0] for i in candarr],c=col)
axarr[0].set_title('scatter plot of the proposed candidates, green indicate accepted, red indicate rejected')
axarr[1].hist([i[0] for i in candarr if i[1]=='g'],bins=100)
axarr[1].set_title('Histogram of the accepted candidates')
plt.show()
```
