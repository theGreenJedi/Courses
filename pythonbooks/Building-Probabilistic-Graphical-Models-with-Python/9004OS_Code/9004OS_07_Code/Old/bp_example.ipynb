In this notebook, we shall use the opengm library(https://github.com/opengm/opengm) to run loopy belief propagation.

OpenGM is a C++ library with Python wrappers that has several implementations of inference algorithms. Please refer to the OpenGM website for installation instructions. For the Python wrapper, Opengm requires installation of HDF5 libraries as well.

We shall use the BP algorithm to perform image segmentation. We first load the image and convert it into a grayscale image.

```python
import opengm
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import Image

fname = 'cow_image.jpg'
image = Image.open(fname).convert("L")
arr = np.asarray(image).astype(float)/255
plt.imshow(arr, cmap = cm.Greys_r)
plt.show()
```

We'll use Opengm's model visualization to better understand the structure of the markov network. In the snippet below, we create a markov network with 9 nodes, arranged in a 3x3 grid. The variables numbered 0-8 are indicated as white unfilled circles.

The grid2d2Order method create a set of factors. The first set of factors are unary potentials associated with each variable, and these are indicated in black squares numbered 0-8 attached to each node.

The second set of factors are pairwise potentials that connect adjacent nodes. These are also indicated in black squares, numbered 9-20.

```python
numLabels=2
shape=(3,3)
unaries=numpy.random.rand(shape[0],shape[1],numLabels)
potts=opengm.PottsFunction([numLabels,numLabels],0.0,0.5)
gm=opengm.grid2d2Order(unaries=unaries,regularizer=potts)
opengm.visualizeGm( gm,plotFunctions=False,
                    plotNonShared=True,relNodeSize=0.9)

```

The network has 21 factors, and the node on the edge (0) has 3 factors associated with it, its unary potential(0), and its pairwise potential (9,10). Similarly the nodes in the middle (4) will have 4 pairwise factors in addition to its unary factor. 

```python
print "number of factors:",gm.numberOfFactors
print "number of factors of node 0: ",gm.numberOfFactorsOfVariable(0)
print "number of factors of node 4: ",gm.numberOfFactorsOfVariable(4)
```

We'll create a markov network for the image in a similar fashion as the 3x3 network above.

First we create a numpy array with the same dimensions of the image. We wish to classify each pixel into one of 2 label classes. For each pixel, we have a additional dimension (of the same length as the class labels), which are initialized with random probabilities (of assignment to each label class).

```python
shape=arr.shape
numLabels=2
unaries=numpy.random.rand(shape[0], shape[1],numLabels)
unaries[:,:,0]=arr 
#Question: why are multiple dimensions required to store class labels? The above line only assigns the image pixel data to the first
#dimension. Assigning it to all the dimensions causes image segmentation to fail.
```

A potts model is similar to an Ising model, which is a Markov network shaped as a grid. Ising models were initially used to model spins of electrons in a grid. The Ising model example of an energy-based model, where the energy is minimized if each electron's spin is in the same direction as its neighbours. In our image segmentation example, the energy is minimized if each pixel's class label is in agreement with the label class of its neighbours.

To quantify how 2 neighbours agree or disagree with each other, we use a distance function. In our case, we use the potts function as our distance function. They return values of 0 and 0.5 if the label assignments agree or do not agree, respectively. 

We use the utility method grid2d2Order, to create the graphical model with unary and pairwise potentials, while mentioning that the factors are composed of Potts functions.

```python
potts=opengm.PottsFunction([numLabels,numLabels],0.0,0.5)
gm=opengm.grid2d2Order(unaries=unaries,regularizer=potts)
```

We are almost ready to start running inference on the markov network. Opengm allows us to create a visitor class that can observe the process of inference. In the 'visit' method which is called at intervals, we can observe the energy minimization in progress. We can also see how the label assignments to each pixel improve visually as the overall energy of the markov net is reduced.

We then create an instance of the BeliefPropagation, and run inference on it.


```python
imgplot=[]

class PyCallback(object):
    def appendLabelVector(self,labelVector):
        labelVector=labelVector.reshape(self.shape)
        imgplot.append([labelVector])
    def __init__(self,shape,numLabels):
        self.shape=shape
        self.numLabels=numLabels
        matplotlib.interactive(True)
    def begin(self,inference):
        print "beginning of inference"
        #assign the initial label vector before start of inference.
        self.appendLabelVector(inference.arg())
    def end(self,inference):
        print "end of inference"
    def visit(self,inference):
        gm=inference.gm()
        #the arg method returns the (class) labeling at each pixel.
        labelVector=inference.arg()
        #evaluate the energy of the graph given the current labeling.
        print "energy  ",gm.evaluate(labelVector)
        self.appendLabelVector(labelVector)


inf=opengm.inference.BeliefPropagation(gm,parameter=opengm.InfParam(damping=0.5))
callback=PyCallback(shape,numLabels)
visitor=inf.pythonVisitor(callback,visitNth=1)

inf.infer(visitor)
```

We view the first 6 class label vectors, and we can see that the cow's shape is easily identified from the labels assigned to each pixel.

```python
fig = plt.figure(figsize=(16, 12))
for (counter, img) in enumerate(imgplot[0:6]):
    a=fig.add_subplot(3,2,counter+1)
    plt.imshow(img[0],cmap=cm.gray, interpolation="nearest")

plt.draw()
```

Finally, we plot the final label assignment for each pixel, which is the MAP (Maximum a posteriori) assignment of class labels to pixels, along with the original image. We can see the result isn't perfect. Indeed, there exists high-performing methods for image segmentation.
