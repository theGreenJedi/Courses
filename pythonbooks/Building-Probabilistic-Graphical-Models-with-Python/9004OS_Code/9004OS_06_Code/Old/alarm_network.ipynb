In this notebook we shall use the alarm network to understand the details of variable elimination.

```python
from libpgm.graphskeleton import GraphSkeleton
from libpgm.nodedata import NodeData
from libpgm.discretebayesiannetwork import DiscreteBayesianNetwork
from libpgm.tablecpdfactor import TableCPDFactor
import itertools
import pandas as pd

nd = NodeData()
skel = GraphSkeleton()
jsonpath="alarm.txt"
nd.load(jsonpath)
skel.load(jsonpath)
skel.toporder()
    
bn = DiscreteBayesianNetwork(skel, nd)
```

Our first attempt to compute marginals will use the joint distribution over all variables, and then sum out the variables we don't care about.

To list the joint distribution, we shall multiply all the factors. The joint distribution computed from the product of factors is listed below, and its complete specification has 32 entries.

```python
#a method that prints the distribution as a table.
def printdist(jd,bn,normalize=False):
    x=[bn.Vdata[i]["vals"] for i in jd.scope]
    zipover=[i/sum(jd.vals) for i in jd.vals] if normalize else jd.vals
    k=[a + [b] for a,b in zip([list(i) for i in itertools.product(*x[::-1])],zipover)]
    df=pd.DataFrame.from_records(k,columns=[i for i in reversed(jd.scope)]+['probability'])
    return df

#instantiate TableCPDs for all the nodes and multiply them. The result is in the factor that calls the methods.
jc=TableCPDFactor("johncalls",bn)    
[jc.multiplyfactor(TableCPDFactor(i,bn)) for i in bn.V if i != "johncalls"]
df=printdist(jc,bn,normalize=True)
print "values in joint distribution ",len(jc.vals)
df
```

We wish to infer the probability that the earthquake occurred and mary and john called ( $ P (earthquake=True \mid johncalls=True,marycalls=True) $, henceforth written as $ P(E\mid jc,mc) $ ). 
Let us first understand how we arrive at the value of the $ P (E\mid jc,mc) $. Using the product rule, $ P(E,jc, mc) =P (E\mid jc,mc) \times P(jc) \times P(mc)  $, we get $ P (E\mid jc,mc) = { P(E,jc, mc)  \over P(jc) P(mc) } $.

From the joint distribution, we can get the desired by performing the following operations. First, we notice that we are interested in data from three columns, earthquake,johncalls, marycalls. So we first collapse the other columns (and the data in it), in an operation called <b>factor marginalization. </b>. 

Throughout the following discussion, we shall use the following shortforms: johncalls (jc), marycalls (mc), alarm (a),burglary(b) and earthquake(e). We shall also overload the notation a bit. Since all the variables are binary, $ P (E\mid jc,mc)$ means that the evidence variables jc and mc are true, unless otherwise mentioned. That is, $ P (E\mid jc,mc)$ is the same as $ P (E=True\mid jc=True,mc=True)$.

Using the pandas dataframe that has the joint distribution, we group only the columns we want, and sum out the values in the probability column.

```python
t2=df.groupby(['earthquake', 'johncalls','marycalls'],as_index=False)
t3=t2['probability'].sum()
t3
```

How does this operation happen? Lets take the 0th row (all values false) as an example and see how the value is obtained.  The value of 0.9368 is obtained by summing up all values of the other columns. We select rows in the table where the values of earthquake, johncalls, marycalls are all 'F'. 

```python
df.loc[(df["earthquake"] == 'F' ) & (df["johncalls"] == 'F') & (df["marycalls"] == 'F'), :]
```

We select the rows from the above table and only choose the probability column, and sum up the values, which turns out to be 0.9368. 

```python
df.loc[(df["earthquake"] == 'F' ) & (df["johncalls"] == 'F') & (df["marycalls"] == 'F'), "probability"].sum()
```

For $ P(E\mid jc,mc) $, we need only rows 0 and 4 from the table above. These rows correspond to $ P(E=T,jc=T,mc=T) $ and  $ P(E=F,jc=T,mc=T) $. This operation is called <b>Factor Reduction</b>, where the other rows are removed because they do not have necessary assignment (that is, values are not true) for johncalls and marycalls. 

```python
t4=t3.loc[ (t3["johncalls"] == 'T') & (t3["marycalls"] == 'T'), :]
t4
```

This is almost the probabilities that we require, except that the probabilities do not sum up to 1. We therefore normalize by dividing each probability by the sum of all probabilities.

```python
psum=t4['probability'].sum()
t4['probability']=t4['probability']/psum
t4
```

What we have shown above is the brute-force method to arrive at the value of a probability query. This method is inefficient for several reasons.

One, it does not use any of the conditional independencies that are present in the model. Lets say we want to know the probability $ P(A\mid B) $, and if nodes X and Y are conditionally independent of A and B,  there is no value in calculating the joint distribution which includes X and Y, and then summing out the values.

Secondly, we already know that eliciting the probability values in the joint distribution is not an easy task, nor is storing and manipulating the large joint distribution. For just 5 binary variables in the alarm example, the joint distribution has 32 rows (or  $ 2^5 $), and for larger networks it becomes impossible to learn, store and manipulate the joint distribution. Even after having the joint distribution, summing out the other values has a time complexity of $ 2^n $. 

We shall now look at the Variable Elimination algorithm which is an efficient method to calculate probabilities. The efficiency results from avoiding duplicate operations as well as avoiding calculation of probabilities that are conditionally independent. This can be used to calculate the marginal probabilities even if any evidence/observed variables exist. 

We shall be use <b>Factors</b> which are similar to the tables we have seen above. The operations of Factor Product (which is a cartesian product between elements of 2 sets), Factor Marginalization and Factor Reductions are the calculus by means of which we can manipulate Factors. 

Each Factor has a scope. If we think of a factor as a table, the scope is the columns associated with the table. The 5 CPDs defined in alarm.txt are the initial factors. The 5 factors are $ \phi (burglary)$ , $  \phi (earthquake) $, $ \phi (alarm,earthquake,burglary) $ (corresponding to the CPD for $ P (Alarm \mid earthquake,burglary) $), $ \phi (johncalls,alarm) $ and  $ \phi (marycalls,alarm) $ .


The probability that we wish to calculate, $ P(E\mid jc,mc) $ should correspond to a factor that has same three variables in its scope, which is $ \phi (E,jc,mc) $. 

To create $ \phi (E,jc,mc) $, we have to use the Factor product operation, where the variables in the scope of the resulting factor should appear in one of the factors being multiplied. For example, the set of factors $  \{ \phi (jc,a) , \phi (mc,a) \} $ has 3 variables in scope $ \{ jc, mc , a \} $. Therefore the product of factors $   \phi (jc,a) , \phi (mc,a) $ will give us a factor $ \phi (A,jc,mc) $, where the probability values are arrived at by multiplying the line where the assignments match.

Lets see an example of how the factor product is calculated. We first print the factors $ \phi (mc,a) $  and $ \phi (jc,a) $ . Although the only argument to the TableCPDFactor is "marycalls", "marycalls" is the node name for the CPD that corresponds $ P(mc \mid A) $.

```python
mc3=TableCPDFactor("marycalls",bn)
printdist(mc3,bn,normalize=False)
```

```python
jhc=TableCPDFactor("johncalls",bn)
printdist(jhc,bn,normalize=False)
```

We can use the libpgm method <i>multiplyfactor</i> to calculate the factor product of $ \phi (mc,a) $  and $ \phi (jc,a) $ .

```python
jhc.multiplyfactor(mc3)
df2=printdist(jhc,bn,normalize=False)
df2
```

We can observe the first entry in the table above (mc=T,a=T,jc=T) has a probability of 0.63. This is obtained by multiplying the first row (with an index 0) in the $ \phi (mc,a) , \phi (jc,a) $ tables. ([a=T,mc=T,0.7] and [a=T,jc=T,0.9] and 0.9 x 0.7 = 0.63). 

To calculate $ \phi (E,jc,mc) $, we use the following steps.

$ P(E,jc,mc)  = \sum_a \sum_b P(E,jc,mc,a,b) $
Where the r.h.s is the full joint distribution, and we sum out the values of a and b.

We can use the independencies to rewrite the r.h.s as

$ \sum_a \sum_b P(E,jc,mc,a,b) = \sum_a \sum_b P(b) P(E) P(a\mid e,b) P(jc\mid a) P(mc\mid a) $

Some of the summing operations can be pushed in

$ \sum_a \sum_b P(E,jc,mc,a,b) = P(E) \sum_b P(b) \sum_a  P(a\mid e,b) P(jc\mid a) P(mc\mid a) $

So we have to deal with the r.h.s using a set of factor operations. The term $ P(E) \sum_b P(b) \sum_a  P(a\mid e,b) P(jc\mid a) P(mc\mid a) $ corresponds the factor $ \phi_1 (E) \sum_b \phi_2 (b) \sum_a  \phi_3 (a\mid e,b) \phi_4 (jc\mid a) \phi_5 (mc\mid a) $

This product of factors slowly collapses by doing a series of factor product and factor marginalizations. 

We start by creating a new factor, the product of $ \phi_6 (a,e,b) = \prod \phi_3 (a\mid e,b) \phi_4 (jc\mid a) \phi_5 (mc\mid a) $

It may seem that we have got the joint distribution again, which is because of the way the alarm network is structured. In all networks, the maximum size of the largest factor will be equal to the minimum 'tree width', which we shall see later in this chapter.


```python
mc3=TableCPDFactor("marycalls",bn)
al=TableCPDFactor("alarm",bn)
phi6=TableCPDFactor("johncalls",bn)

[phi6.multiplyfactor(i) for i in [mc3,al]]
printdist(phi6,bn)
```

After the first factor product operations, we are left with $ \phi_1 (E) \sum_b \phi_2 (b) \sum_a  \phi_6 (a,e,jc,mc)  $.  We now have to sum out the variable A. 

```python
phi6.sumout("alarm")
printdist(phi6,bn)
```

Summing out the alarm variable gives us $ \phi_1 (E) \sum_b \phi_2 (b) \phi_7 (e,jc,mc)  $. We next do a factor product of $ \phi_2 (b) \phi_7 (e,jc,mc)  $. 

```python
phi8=TableCPDFactor("burglary",bn)
phi8.multiplyfactor(phi6)
printdist(phi8,bn)
```

In the resulting factor $ \phi_8 (b,jc,mc) $, we sum out the burglary variable. 

```python
phi8.sumout("burglary")
phi9=phi8
printdist(phi9,bn)
```

We are now left with a new factor, $ \phi_9 (e,jc,mc) $, and to get the result, we only need to get the product of $ \phi_1 (E)  \phi_9 (e,jc,mc)   $.

```python
phi10=TableCPDFactor("earthquake",bn)
phi10.multiplyfactor(phi9)
printdist(phi10,bn)
```

This is the fully specified CPD for the 3 variables. However, we are only interested in $ P(E\mid jc,mc) $, which are the rows for which jc and mc have 'T', those are rows 0 and 1. We can use the reducefactor method to do the same.

```python
phi10.reducefactor("marycalls",'T')
phi10.reducefactor("johncalls",'T')
printdist(phi10,bn)
```

We just have to normalize the values of the probabilities to get the final result, which matches the result we obtained from summing out the excess variables in the joint distribution.

```python
summ = sum(phi10.vals)
phi10.vals=[i/float(summ) for i in phi10.vals]
printdist(phi10,bn)
```

We use libpgm's implementation of variable elimination to get the results straightaway.

```python
from libpgm.tablecpdfactorization import TableCPDFactorization
nd = NodeData()
skel = GraphSkeleton()
jsonpath="alarm.txt"
nd.load(jsonpath)
skel.load(jsonpath)
skel.toporder()
    
bn = DiscreteBayesianNetwork(skel, nd)
evidence = {"johncalls":'T', "marycalls":'T'}
query = {"earthquake":'T'}

fn = TableCPDFactorization(bn)
result = fn.condprobve(query, evidence)
```

```python
printdist(result,bn)
```
