#Getting Ready

In this recipe we're going to look more in depth at what it takes to tune a Decision Tree Classifier.  There are several options and we only looked at one way to tune in the last recipe.  We'll fit a basic model, and actually look at what the decision tree looks like.  Then we'll re-examine after each decision and point out how various changes have influenced the structure.

If you want to follow along in this recipe you'll need to install pydot.

#How to do it

```python
from sklearn import datasets
X, y = datasets.make_classification(1000, 20, n_informative=3)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(X, y)
```

Ok, so now we that we have a stock classifier fit, we can view it quite simply.

```python
from StringIO import StringIO
from sklearn import tree
import pydot 
```

```python
str_buffer = StringIO()
tree.export_graphviz(dt, out_file=str_buffer)
graph = pydot.graph_from_dot_data(str_buffer.getvalue())
graph.write_jpg("978-1-78398-948-5_05_04.png")
```

```python
Image("978-1-78398-948-5_05_04.png")
```

Wow that is a very complex tree.  It would most likely overfit the data.  First let's reduce the max depth.

```python
dt = DecisionTreeClassifier(max_depth=5)
dt.fit(X, y); 
```

As an aside, if you're wondering why the semicolon, by default the repr will show, which for a decision tree is the model.  For example, fit acttually returns the decision tree object.  This allows for chaining.

```python
dt = DecisionTreeClassifier(max_depth=5).fit(X, y)
```

Now, back to the regularly scheduled program.

Also because we're going to do this plotting a few times, let's create a function.

```python
def plot_dt(model, filename):
    
    str_buffer = StringIO()
    tree.export_graphviz(model, out_file=str_buffer)
    
    graph = pydot.graph_from_dot_data(str_buffer.getvalue())
    graph.write_jpg(filename)
```

```python
filename = "978-1-78398-948-5_05_05.png"
plot_dt(dt, filename)
Image(filename)
```

That's a much simpler tree.  Let's look at what happens when we use entropy as the splitting criteria.

```python
dt = DecisionTreeClassifier(criterion='entropy', max_depth=5).fit(X, y)
```

```python
filename="978-1-78398-948-5_05_06.png"
plot_dt(dt, filename)
Image(filename)
```

It's good to see that the first two splits are the same features, and the first few after that are interspersed with similar amounts.  That's a good sanity check.

Also, can you see how entropy for the first split is 0.999, but for the first split when using the gini coefficient is 0.5.  This has to do with the how the two measure the split of a Decision Tree.  See the How it works section for more info.

Let's have a look at one more, here it'll be the min_samples_leaf.  This controls how many datapoints, at minimum, are required to create a leaf.

```python
dt = DecisionTreeClassifier(min_samples_leaf=10, criterion='entropy', max_depth=5).fit(X, y)
```

```python
filename = "978-1-78398-948-5_05_06.png"
plot_dt(dt, filename)
Image(filename)
```
