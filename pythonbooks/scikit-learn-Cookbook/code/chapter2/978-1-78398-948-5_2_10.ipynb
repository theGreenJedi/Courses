```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#Getting Ready

SVM is often thought of as a means for classification, however it can be used for regression.  So instead of fitting a hyperplane to classify the data, we'll predict a continuous value given some inputs.

#How to do it

We'll look a simple case where we have a single predictor so we can look at how the choice of radial basis function affects the output.

First let's create a trickier dataset, we'll use a sine function plus noise to generate the data.

```python
import numpy as np
X = np.linspace(-4*np.pi, 4*np.pi, 500).reshape(500, 1)
y = 2*sin(X + .001) + np.random.normal(0, 1, (500, 1))
```

```python
f, ax = plt.subplots(figsize=(7, 5))
ax.scatter(X, y)
ax.set_title("Tricky Regression Data")
f.savefig("tricky_regression.png")
```

If we were to use Linear Regression to the dataset it would like like:

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
preds = lr.predict(X)
```

```python
f, ax = plt.subplots(figsize=(7, 5))
ax.scatter(X, y, alpha=.25)
ax.plot(X, preds, lw=5, alpha=.5, label='Linear Regression')
ax.set_title("Tricky Regression Data with Regression Line")
ax.legend()
f.savefig("Tricky_Regression_Line.png")
```

Now if we ere to run the regression, but instead applied SVM regression with a radial basis function.  RBF is the default kernal for SVR, so no modifications are needed.

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X, y.ravel())
y_preds = svr.predict(X)
```

```python
f, ax = plt.subplots(figsize=(7, 5))
ax.scatter(X, y, alpha=.25)
ax.plot(X, preds, lw=5, alpha=.95, label='Linear Regression')
ax.plot(X, y_preds, lw=5, alpha=.95, label='SVR RBF Kernel')
ax.set_title("Tricky Regression Data with Regression Line")
ax.legend()
f.savefig("tricky_regression_svr.png")
```

#How it works

Support Vector Machines and by extension Support Vector Regression is a very deep subject.  There are a few important things to call out.

First, SVR uses a kernel function to map the data onto a new space.  This is the same idea we've seen before with KernelPCA.

Second, SVR doesn't use the entire dataset to fit the model, rather it uses the closest datasets to the decision boundery to fit the regression.  This is a desireable quality, because it can help reduce overfitting.

#There's more

It's actually possible to use a linear kernel function.  Notice the subtle difference in how the linear regression compares to SVR with a linear kernel.

```python
from sklearn.svm import SVR
svr = SVR('linear')
svr.fit(X, y.ravel())
y_preds = svr.predict(X)
```

```python
f, ax = plt.subplots(figsize=(7, 5))
ax.scatter(X, y, alpha=.25)
ax.plot(X, preds, lw=5, alpha=.95, label='Linear Regression')
ax.plot(X, y_preds, lw=5, alpha=.95, label='SVR Linear Kernel')
ax.set_title("Tricky Regression Data with Regression Line")
ax.legend()
f.savefig("linear_kernel_svr.png")
```
