#Getting Ready

In the last chapter we discussed Gaussian Mixture Models.  These models let us relax our assumptions regarding the variances of the clusters.

A Dirichlet Process GMM allows us to relax the assumptions regarding a predefined number of clusters.  Instead we let the number of clusters be a random variable that is disbtibuted according to a Dirichlet distribution.

The Dirichlet distribution comes up often in this type of situation.  For example, with Latent Dirichlet Allocation the prior the topic of a document might be set to be a Dirichlet.

#How to do it

DPGMM api is very similar to GMM.

```python
from sklearn.datasets import make_blobs
blobs, labels = make_blobs()
```

```python
from sklearn.mixture import DPGMM
dpgmm = DPGMM()
```

```python
dpgmm.fit(blobs)
```

```python
#the mean(s) of the distribution
dpgmm.means_
```

```python
#the precision (inverse of the variance)
dpgmm.precs_
```

Since Dirichlet distributions often are used the Bayesian sense, the precision is used as opposed to the variance.

#How it works

You might have noticed that we didn't get a very good fit.  The DPGMM fit a single cluster whereas the data generated has 3.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
f, ax = plt.subplots(figsize=(7, 5))

ax.set_title("Plots")

ax.scatter(blobs[:, 0], blobs[:, 1], label="Blobs", c=labels)
ax.scatter(dpgmm.means_[:, 0], dpgmm.means_[:, 1], s=100, label='Center', color='c')
ax.legend()
```

Let's alter the number of components to see if that might affect the fit.

```python
dpgmm = DPGMM(n_components=3)
dpgmm.fit(blobs)
```

```python
f, ax = plt.subplots(figsize=(7, 5))

ax.set_title("Plots")

ax.scatter(blobs[:, 0], blobs[:, 1], label="Blobs", c=labels)
ax.scatter(dpgmm.means_[:, 0], dpgmm.means_[:, 1], s=100, label='Center', color='c')
ax.legend()
```

Clearly this looks better, but deliniating between the red and blue cluster would probably be very difficult for a human to do if the plots weren't labeled.
