```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=300, centers=4,
                  random_state=0, cluster_std=1.0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='gray');
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

$$ J = \sum_k \sum_i \lvert x_i - \mu_k \rvert^2 $$

The *distortion* for the $k^{th}$ cluster is this part

$$ \lvert x_i - \mu_k \rvert^2 $$

Each cluster's distortion is equal to the sum of the squared distances between its centroid and its constituent instances. The distortion is small for compact clusters and large for clusters that contain scattered instances. In other words, the distortion is a measure of the compact-ness of a particular cluster.

```python
m_distortions=[]
for k in range(1,7):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    m_distortions.append(sum(np.min(cdist(X, kmeans.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
plot(m_distortions,'-o',ms=10)
```

```python
def draw_distortion(X):    
    m_distortions=[]
    sscore=[]
    for k in range(2,7):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(X)
        sscore.append(silhouette_score(X,kmeans.labels_))
        m_distortions.append(sum(np.min(cdist(X, kmeans.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
    plot(range(2,7),m_distortions,'-o',ms=10,label='distortion')
    plot(range(2,7),sscore,'-s',ms=10,label='silhouette')
    legend(loc=0,fontsize=18)
draw_distortion(X)
```
