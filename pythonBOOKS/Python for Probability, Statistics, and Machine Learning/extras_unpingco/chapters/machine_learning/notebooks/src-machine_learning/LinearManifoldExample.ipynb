```python
from __future__ import division
d=3  # dimension
n=10
X = np.matrix(rand(n,d)) # choose any set of vectors
```

```python
wo = np.matrix(arange(d)).T 
y = X*wo + np.random.randn(X.shape[0],1)

# training weights
w = linalg.inv(X.T*X)*X.T*y
h = X*w
Ein=np.linalg.norm(h-y)**2

# out of sample error
yp = X*wo + np.random.randn(X.shape[0],1)
Eout=np.linalg.norm(h-yp)**2
```

```python
def est_errors(d=3,n=10,niter=100):
    assert n>d
    wo = np.matrix(arange(d)).T 
    Ein = list()
    Eout = list()
    X = np.matrix(rand(n,d)) # choose any set of vectors
    for ni in xrange(niter):
        y = X*wo + np.random.randn(X.shape[0],1)
        # training weights
        w = linalg.inv(X.T*X)*X.T*y
        h = X*w
        Ein.append(np.linalg.norm(h-y)**2)
        # out of sample error
        yp = X*wo + np.random.randn(X.shape[0],1)
        Eout.append(np.linalg.norm(h-yp)**2)
    return (np.mean(Ein)/n,np.mean(Eout)/n)
```

```python
# %qtconsole
```

```python
d=18
xi = arange(d*2,d*10,d//2)
ei,eo=np.array([est_errors(d=d,n=n,niter=100) for n in xi]).T
```

```python
plot(xi,ei,'or',label='$\hat{E}_{in}$')
plot(xi,eo,'bo',label='$\hat{E}_{out}$')
plot(xi,(1-d/array(xi)),label='${E}_{in}$',color='r')
plot(xi,(1+d/array(xi)),label='${E}_{out}$',color='b')
hlines(1,xi.min(),xi.max(),lw=3)
xlabel('size of training set ($n$)',fontsize=22)
ylabel('MSE',fontsize=22)
legend(loc=(1,0),fontsize=18)
title('dimension = %d'%(d))
```

```python
def draw_plot(d=8):
    xi = arange(d+1,d*10,d//2)
    ei,eo=np.array([est_errors(d=d,n=n,niter=d*10) for n in xi]).T
    plot(xi,ei,'or',label='$\hat{E}_{in}$')
    plot(xi,eo,'bo',label='$\hat{E}_{out}$')
    plot(xi,(1-d/array(xi)),label='${E}_{in}$',color='r')
    plot(xi,(1+d/array(xi)),label='${E}_{out}$',color='b')
    hlines(1,xi.min(),xi.max(),lw=3)
    xlabel('size of training set ($n$)',fontsize=22)
    ylabel('MSE',fontsize=22)
    legend(loc=(1,0),fontsize=18)
    title('dimension = %d'%(d))
    axis(ymax=2,ymin=0)
```
