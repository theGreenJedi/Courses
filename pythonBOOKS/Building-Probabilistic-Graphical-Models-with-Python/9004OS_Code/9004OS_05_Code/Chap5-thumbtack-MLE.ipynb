In this notebook, we shall examine how to learn the parameters of a thumbtack using Maximum likelihood estimation.

We first import the bernoulli distribution. Working backwards, we set the bias of the thumbtack(by choosing a value for the parameter theta), and then flip the thumbtack a number of times(by sampling random values)


```python
from scipy.stats import bernoulli
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

param_theta=0.3
num_flips=100
```

```python
bernoulli.rvs(0.3,size=10)
```
