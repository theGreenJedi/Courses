In this notebook, we shall use Bayesian methods to determine the bias of the thumbtack, which we shall call parameter theta.

We shall use the excellent PyMC library for the same. Please install PyMC using "easy_install PyMC". 

We shall avoid going into the finer details of PyMC. Excellent documentation for PyMC can be found both in the PyMC docs as well as the book "Probabilistic Programming and Bayesian methods for Hackers" by Cam Davidson-Pilon's, which can be found online at http://camdavidsonpilon.github.io/Probabilistic-Programming-and-Bayesian-Methods-for-Hackers/.

For estimating theta, we shall first generate 30 samples from the Bernoulli distribution, where the controlling parameter has a value of 0.3 (which would correspond to roughly 30% of the generated values being 1, and the rest being 0). 

```python
from pymc import *
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
from pymc.Matplot import plot

data=bernoulli.rvs(0.3,size=30)
```

We attempt to find the parameters of the posterior distribution, and we've learnt that the posterior depends on the prior, as well the data. There are many choices of prior available to us, in this case we shall use the uniform prior. Since we have little idea about the bias of the thumbtack, we believe it can lie anywhere between 0 and 1. 

We already have observed data from flipping the thumbtack (saved in the 'data' variable), and we provide that to the Bernoulli object. We also tell the Bernoulli object that it has a parent, which is the uniform prior. We then create a Model, the nodes in the model are the Uniform and Bernoulli objects, which have a parent-child relationship.

PyMC has various kinds of variables, such as stochastic (for example, the uni_prior variable, which can take different values based on the parameter theta), and deterministic (such as 'bern', whose values are decided by its parents). Finally, all the variables in the model are wrapped in a Model object. 
 
For all variables where observed is not True, PyMC's simulations will tickle the value of the variable during the simulation and the value (of uni_prior in our case) will start to approximate its posterior values.

```python
#create a uniform prior, the lower and upper limits of which are 0 and 1
uni_prior = Uniform('uni_prior', lower=0.1,upper=0.9 )

bern = Bernoulli('bern',p=uni_prior, value=data,observed=True)
model=Model([uni_prior,bern])
```

Samples from the Prior disttribution show that the prior is uniformly distributed between 0 and 1 on the x axis.

```python
samples = [uni_prior.random() for i in range(20000)]
plt.hist(samples, bins=100, normed=True )
plt.title("Prior distribution for $\ theta$")
plt.show()
```

In the following snippet, we use a sampling method called Markov Chain Monte Carlo (hence MCMC), and then provide the model we have created. MCMC is a method to draw samples from the posterior distribution. We draw 5k samples, and draw a histogram of the samples (called traces in MCMC parlance). 

Bayesian statistic will represent the parameter theta as a random variable, and we can see from the plot below that the distribution peaks around the 0.3 mark, which hews to the same distribution that we drew the samples from. The peak of the distribution hill is called a point estimate in Bayesian parlance, which if the value we choose if we want to display the parameter theta as a single value and not as a distribution.

PyMC's default plot method plots the trace of each unobserved variable (the prior in our case), the posterior distribution of the uni_prior (we can see that it has been modified from the uniform prior to something that approximates a normal distribution with a peak near 0.4), and the autocorrelation of the same variable.

```python
mc1 = MCMC(model)
mc1.sample(iter=5000,burn=1000)
    
plot(mc1)
plt.show()
```

It can be seen from the histogram of the posterior distribution of uni_prior that the peak (around 0.43) is quite different from the ground truth (0.3). We had used only 30 samples for the initial model, can increasing the number of samples help improve the distribution?

```python
bern = Bernoulli('bern',p=uni_prior, value=bernoulli.rvs(0.3,size=5000),observed=True)

mc2 = MCMC(Model([uni_prior,bern]))
mc2.sample(iter=5000,burn=1000)
    
plot(mc2)
plt.show()
```

It apparently does, as the black line indicates a mean of 0.3. 

We can also use PyMC's graphing capabilities to create a image of the hierarchical model. We need to install a few python libraries, such as pydot, graphviz (and install base graphviz as well).

Creating the dot-format graph representation is a simple one-liner.

```python
gdot=pymc.graph.dag(mc2)
```

We can try to see the effect of increasing the amount of data available to the model

```python
num_samples=[20,50,100,500]

def getModels(num_samples):
    b=[Bernoulli('bern'+str(i),p=uni_prior, value=bernoulli.rvs(0.3,size=i),observed=True) for i in num_samples]
    return [Model([uni_prior,i]) for i in b]

def runsamp(mod):
    mc = MCMC(mod)
    mc.sample(iter=5000,burn=300)
    return mc

models=[runsamp(i) for i in getModels(num_samples)]
```

```python
#pymc.Matplot.autocorrelation(models[0])

for i in models:
    pymc.Matplot.histogram(i)
```

```python
f, axarr = plt.subplots(len(models), 1)
f.tight_layout()
plt.subplots_adjust(hspace = 1)

for i, mod in enumerate(models):
    ax1 = plt.subplot(len(models),1,i+1)
    ax1.hist(mod.trace('uni_prior')[:],bins=30)

    msg="number of samples ",num_samples[i]
    
    ax1.set_title(msg)
    ax1.set_xlabel("$theta$")

    #ax1.set_ylim([0,0.8])
    
plt.show()
```
