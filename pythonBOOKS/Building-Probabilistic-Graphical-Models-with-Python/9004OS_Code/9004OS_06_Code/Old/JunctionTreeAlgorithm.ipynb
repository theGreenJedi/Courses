In this notebook, we shall use the Bayesian Belief Network (BBN) library to run exact inference using the junction tree algorithm. The library is available here (https://github.com/eBay/bayesian-belief-networks) and the documentation to install the library is mentioned on the Github page.

BBN has functionality to load networks stored in a the Bayesian Interchange Format (shortenend to bif), which is developed by the Bayesian community to foster easier data sharing between different inference tools. 

We shall load the "asia" network (http://www.bnlearn.com/bnrepository/#asia), which is a toy bayesian network used in patient diagnosis. The network tries to elicit the possibility of lung cancer or tubercolosis based on factors such as travel to Asia, smoking history, xray results and so on. Further details on the asia network can be found here (http://www.norsys.com/tutorials/netica/secA/tut_A1.htm). 

After the mandatory imports, we parse the .bif format file with the bif_parser module, which returns a bayes net object.

```python
import bif_parser
import prettytable
import pydot
from IPython.core.display import Image 
from bayesian.bbn import *

name = 'asia'

module_name = bif_parser.parse(name)
module = __import__(module_name)
bg = module.create_bbn()
```

We can view the bayes net using the graphviz functionality offered by BBN. 

```python
def show_graphgiz_image(graphviz_data):
    graph = pydot.graph_from_dot_data(graphviz_data)
    graph.write_png('temp.png')
    return 'temp.png'

sf=bg.get_graphviz_source()
Image(filename=show_graphgiz_image(sf))
```

The next snippet shows the moralization phase. Notice that the V-structures (for example f_tub -> f_either <-f_lung), have their parents joined with a new link.

```python
gu=make_undirected_copy(bg)
m1=make_moralized_copy(gu,bg)
s2=m1.get_graphviz_source()
Image(filename=show_graphgiz_image(s2)) 
```

In the next phase, the undirected graph is triangulated, which is the addition of edges between nodes if there is cycle greater than length 4. 
Notice that f_bronc,f_either,f_lung,f_smoke form a cycle with 4 nodes. In the next section, a link is added between f_either and f_smoke to triangulate it.

```python
cliques, elimination_ordering = triangulate(m1, priority_func)
s2=m1.get_graphviz_source()
Image(filename=show_graphgiz_image(s2)) 
```

Next, we create the clique and the sepsets, which is done completely within the build_join_tree method. The method create cliques according the graph above, and creates the sepsets which are the intersection between each pair of cliques. 

Note that the naming convention for the nodes in the generated graph is not exactly pretty. The clique "Clique_EITHERLUNGTUB" has the variables either, Lung, Tub all smashed together. 

```python
jt=bg.build_join_tree()
sf=jt.get_graphviz_source()

Image(filename=show_graphgiz_image(sf)) 
```

We can see from the variable names that the sepsets contain the intersection of variables from the cliques that they connect. For example, the variables in the Clique (BRONC,Either,Smoke) and (BRONC,DYSP,EITHER) are the variables in the sepset (BRONC,EITHER).

The next step is to create the initial cluster (which are usually the cliques) and initialize potentials.

```python
assignments = jt.assign_clusters(bg)
jt.initialize_potentials(assignments,bg)
```

In the next stage which is the message passing stage, there are 2 messages sent between neighbouring clusters, one in a forward and another in a reverse pass. The details of the message passing are contained within the propagate() method in the junction tree object. The sequence of messages in the network is shown in the following figure. 

```python
Image(filename="/home/kiran/Dropbox/book/chapter6/images/9004OS_06_05.png")
```

So what exactly happens when a message is passed? There are 3 actors in a message pass, the source cluster, the intervening sepset, and the destination cluster. Each of these have a set of potentials (very similar to a CPD along with the associated probabilities, except that it need not be a valid probability distribution). 

For example a message from $\phi_x (bronc, either, smoke)  $  to $ \phi_y (either, lung, smoke) $ would pass the sepset $ \phi_r (either, smoke) $ , where the variables are the scope of each cluster or sepset. 

The message would modify the potentials at $ \phi_r $ and $\phi_y$. 

The first assignment to the sepset R is the potentials in the source cluster with the variables not in the sepset, marginalized out. In the above example, it is $\phi x $ with bronc summed out.

$\phi_R = \sum_{X \setminus R} \phi_x $

The second assignment to the cluster Y 

$\phi_Y = \phi_Y { \phi_R \over \phi_R^{old} } $ 

Lets run the message passing bit using the propagate method.



```python
jt.propagate()
```

When done with all the message passing, we are left with a tree that has consistent beliefs in all its clusters.

For querying the marginal probability for a particular variable (say Bronc), we just have to find a cluster(or sepset) that has Bronc in its scope, and marginalize the other variables.

We can observe that the potentials is a just a table with assignments.

```python
bronc_clust=[i for i in jt.clique_nodes for v in i.variable_names if v =='bronc']
bronc_clust[0].potential_tt
```

We can see that the clique listed above has 'bronc' in its scope, lets try to find the marginal for bronc by marginalizing the dysp and either variables using the first cluster we found (with bronc in scope).

```python
pot=bronc_clust[0].potential_tt

#a function to return the sum for a specific assignment, such as 'bronc,yes'
sum_assignments=lambda imap,tup:sum([v for k,v in imap.iteritems() for i in k if i == tup])

#get the sum for bronc=yes and bronc=no
yes,no=[sum_assignments(pot,('bronc',i)) for i in ['yes','no']]

print 'bronc: yes ', yes/float(yes+no)," no ", no/float(yes+no)
```

Since we claim that the junction tree is consistent, will other clusters containing bronc return the same marginals? Lets use the second cluster with bronc in its scope and marginalize on its potentials.

```python
pot2=bronc_clust[1].potential_tt

#get the sum for bronc=yes and bronc=no
yes,no=[sum_assignments(pot2,('bronc',i)) for i in ['yes','no']]

print 'bronc: yes ', yes/float(yes+no)," no ", no/float(yes+no)
```

Indeed, the values are the same. 

We can use this algorithm to infer the probability of a node with (other) observed variables as well.
