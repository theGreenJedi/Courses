In this notebook, we shall look at the sampling process. We'll first understand what it means to 'arrive at the stationary distribution' for a discrete distribution, and what methods we can use to get there faster.

We'll use the familiar job interview example to anchor the discussion.

The Job interview network has 5 binary-valued variables, which means the joint distribution has 48 rows (2x2x3x3x2, the number of values each variable takes). We are interested in a marginal distribution over a subset of variables, and we have some observed evidence too. 

```python
from libpgm.graphskeleton import GraphSkeleton
from libpgm.nodedata import NodeData
from libpgm.discretebayesiannetwork import DiscreteBayesianNetwork
from libpgm.tablecpdfactorization import TableCPDFactorization
from libpgm.sampleaggregator import SampleAggregator
from libpgm.pgmlearner import PGMLearner
import itertools
import pandas as pd 
import json

def getTableCPD():
    nd = NodeData()
    skel = GraphSkeleton()
    jsonpath="job_interview.txt"
    nd.load(jsonpath)
    skel.load(jsonpath)
    skel.toporder()
    # load bayesian network
    bn = DiscreteBayesianNetwork(skel, nd)
    tablecpd=TableCPDFactorization(bn)
    return tablecpd,bn,skel

#a method that prints the distribution as a table.
def printdist(jd,bn,normalize=False):
    x=[bn.Vdata[i]["vals"] for i in jd.scope]
    zipover=[i/sum(jd.vals) for i in jd.vals] if normalize else jd.vals
    #creates the cartesian product
    k=[a + [b] for a,b in zip([list(i) for i in itertools.product(*x[::-1])],zipover)]
    df=pd.DataFrame.from_records(k,columns=[i for i in reversed(jd.scope)]+['probability'])
    return df
```

We are interested in the marginal probability $ P(Offer,Grades,Interview \mid Admission,Experience) $, where we have observed the value of Admission and Experience.

If we knew nothing about the probabilities in the CPD below, we might want to distribute the probability mass uniformly across all 12 rows, each row getting 1/12th the probability value. Our  'uniform prior' when we start, is seen in the table below.

```python
tcpd,bn,skel=getTableCPD()
query={'Offer':'0','Grades':'0','Interview':'0'}
evidence={'Admission':'0','Experience':'0'}
fac=tcpd.condprobve(query,evidence)
uniform_vals=[1/float(len(fac.vals)) for i in fac.vals]
df=printdist(fac,bn)
df['uniform']=uniform_vals
df
```

To get the desired distribution $ P(Offer,Grades,Interview∣Admission=0,Experience=0) $, we first have to draw samples, reject those that do not satisfy the evidence. 


```python
tcpd,bn=getTableCPD()

def get_vals(sample):
    return tuple([sample[i] for i in [u'Admission', u'Experience', u'Offer', u'Interview',u'Grades']])

#collate the number of samples for each assignment.
def collate_samples(samples):
    m={}
    for k in samples:
        vals=get_vals(k)
        val=m.get(vals,0)
        m[vals]=val+1
    d=[list(k)+[v] for k,v in m.iteritems()]
    df=pd.DataFrame(d,columns=[u'Admission', u'Experience', u'Offer', u'Interview',u'Grades','counts'])
    return df

df=collate_samples(bn.randomsample(1000))
#convert the map into tuples

print "num rows ",df.shape[0]
df.head(10)         
```

Notice that after drawing 1000 samples, we've drawn only 38 unique assignments (out of 48). If we reduce for evidence (Admission=0,Experience=0), we have even fewer samples. How many exactly? 

```python
#create a new data frame with specific assignments to Admission and Experience.
wanted_cols=[u'Offer', u'Interview',u'Grades','counts']

def reduce_evidence(dataf):
    df1=dataf.loc[(dataf['Admission'] == '0') & (dataf['Experience'] == '0'), :]
    return df1[wanted_cols].sort(columns=wanted_cols)

df1=reduce_evidence(df)
print "counts with other samples rejected ", df1['counts'].sum()
df1
```

So only about 32% of our previous drawn samples were useful, and we haven't got any samples for some of the rows. In distributions where the probabilities are low, drawing samples from the required distribution using random sampling is simply too inefficient. 

So far we have been thinking of sampling as a stochastic procedure, where each sample is independent of the previous sample(s).

We can think of each of the 12 rows in the CPD as an individual state, and we can choose specific ways in which we can "move" from one state to another.
