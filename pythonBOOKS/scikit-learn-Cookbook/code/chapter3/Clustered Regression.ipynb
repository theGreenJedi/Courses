#Getting Ready

Our old friend Regression can be used in the context of clustering.  Regression is obviously a supervised technique, and so we'll use KNN clustering rather than KMeans.

For KNN regression we'll use the k closest points in the feature space to build the regression rather than use the entire space like regular regression.

For this recipe we'll use the iris dataset.  You can image if we want to predict something like the petal width for each flower, first clusering by iris species could potentially give us better results.  KNN Regression won't cluster by the species, but we'll work under the assumption that for the same species the Xs will be close, or here the petal length.

#How to do it

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris.feature_names
```

```python
import numpy as np
X = iris.data[:, :2]
y = iris.data[:, 3]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
```

```python
print "The MSE is: {:.2}".format(np.power(y - lr.predict(X), 2).mean())
```

Now for KNNRegression

```python
from sklearn.neighbors import KNeighborsRegressor 
```

```python
knnr = KNeighborsRegressor(n_neighbors=10)
```

```python
knnr.fit(X, y)
```

```python
print "The MSE is: {:.2}".format(np.power(y - knnr.predict(X), 2).mean())
```

Let's look at what k-NN Regression is doing when we tell it to use the closest 10 points for regression.

```python
f, ax = plt.subplots(figsize=(7, 5))

ax.set_title("Sepal Measurements Sized by Petal Width")
ax.scatter(X[:, 0], X[:, 1], s=y*80)
```

```python
f, ax = plt.subplots(nrows=2, figsize=(7, 10))

ax[0].set_title("Predictions")

ax[0].scatter(X[:, 0], X[:, 1], s=lr.predict(X)*80, label='LR Predictions', color='c', edgecolors='black')
ax[1].scatter(X[:, 0], X[:, 1], s=knnr.predict(X)*80, label='k-NN Predictions', color='m', edgecolors='black')

ax[0].legend()
ax[1].legend()

#f.savefig("978-1-78398-948-5_15.jpg")
```

It might be completely clear the predictions are close for the most part, but let's look at the predictions for the Setosa species compared to the actuals. 

```python
setosa_idx = np.where(iris.target_names=='setosa')
setosa_mask = iris.target == setosa_idx[0]
```

```python
# the acutals
```

```python
y[setosa_mask][:5]
```

```python
knnr.predict(X)[setosa_mask][:5]
```

```python
lr.predict(X)[setosa_mask][:5]
```

Looking at the plots again, the setosa species (upper left cluster) was largely overpredicted by linear regression and k-NN is fairly close to the actual values.

#How it works

k-NN Regression very simply by taking the average the the k closest point to the point being tested.

Let's manually predict a single point.

```python
example_point = X[0]
```

Now we need to get the 10 closes points to our example_point.

```python
from sklearn.metrics import pairwise
```

```python
distances_to_example = pairwise.pairwise_distances(X)[0]
```

```python
ten_closest_points = X[np.argsort(distances_to_example)][:10]
ten_closest_y = y[np.argsort(distances_to_example)][:10]
```

```python
ten_closest_y.mean()
```

We can see that our prediction from the k-NN regression matches what we found when we manually found the means of the 10 closest points.
