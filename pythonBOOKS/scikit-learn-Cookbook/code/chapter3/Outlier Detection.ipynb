#Getting Started

In the recipe we'll use KMeans to do outlier detection on a cluster of points.  Upfront it's important to note that there are many "camps" when it comes to outliers and outlier detection.  One the one hand be removing outliers we're potentially removing points that were, in fact, generated by the data generating process.  On the other hand, the case can be made that outliers are due to a measurement error or some other outside factor.

That's the most credence we'll give to the debate, the rest of this recipe is about finding outliers, and we'll work under the assumption that our choice to remove outliers is justified.

The act of outlier detection is a matter of finding the centroids of the clusters then identifying points which are potential outliers by their distances from the centroid.

#How to do it

First we'll generate a single blob of 100 points then identify the 5 points which are furthest from the centroid.  These are the potential outliers.

```python
from sklearn.datasets import make_blobs
import numpy as np
X, labels = make_blobs(100, centers=1)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=1)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
f, ax = plt.subplots(figsize=(7, 5))

ax.set_title("Blob")
ax.scatter(X[:, 0], X[:, 1], label='Points')
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], label='Centroid', color='r')
ax.legend()
#f.savefig("978-1-78398-948-5_12.jpg")
```

Now let's identify the 5 closest points.

```python
distances = kmeans.transform(X)

# argsort returns an array of indexes which will sort the array in ascending order
# so we reverse it via [::-1] and take the top five with [:5]
sorted_idx = np.argsort(distances.ravel())[::-1][:5]
```

```python
f, ax = plt.subplots(figsize=(7, 5))

ax.set_title("Single Cluster")
ax.scatter(X[:, 0], X[:, 1], label='Points')
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], label='Centroid', color='r')
ax.scatter(X[sorted_idx][:, 0], X[sorted_idx][:, 1], label='Extreme Value', edgecolors='g', facecolors='none', s=100)
ax.legend(loc='best')
#f.savefig("978-1-78398-948-5_13.jpg")
```

It's easy then to remove those points if we'd like.

```python
new_X = np.delete(X, sorted_idx, axis=0)
```

Also the centroid as clearly changed with the removal of those points.

```python
new_kmeans = KMeans(n_clusters=1)
new_kmeans.fit(new_X)
```

```python
f, ax = plt.subplots(figsize=(7, 5))

ax.set_title("Extreme Values Removed")
ax.scatter(new_X[:, 0], new_X[:, 1], label='Pruned Points')
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], label='Old Centroid', color='r', s=80, alpha=.5)
ax.scatter(new_kmeans.cluster_centers_[:, 0], new_kmeans.cluster_centers_[:, 1], label='New Centroid', color='m', s=80, alpha=.5)
#ax.scatter(X[sorted_idx][:, 0], X[sorted_idx][:, 1], label='Extreme Value', edgecolors='g', facecolors='none', s=100)
ax.legend(loc='best')
#f.savefig("978-1-78398-948-5_14.jpg")
```

Clearly the centroid hasn't moved much, which is to be expected when only removing the 5 most extreme values.  This process can be repeated until the practioner is satisfied the data is representative of the process.

#How it works

Like we've already seen there is a fundemental connection between the Gaussian distribution and KMeans clustering.  Let's create an empricial Gaussian based off the centroid and the sample covariance matrix and look at the probability of each point.  Theoretically the 5 points we removed 

```python
from scipy import stats
emp_dist = stats.multivariate_normal(kmeans.cluster_centers_.ravel())
```

```python
lowest_prob_idx = np.argsort(emp_dist.pdf(X))[:5]
```

```python
np.all(X[sorted_idx] == X[lowest_prob_idx])
```

So what we've done is found the points which are the least likely given the centroid.  It's worth noting that we assumed that the covariance is eI (MATHIFY) as opposed to the more general S.
