#Getting Ready

We're going to have some fun in this chapter.  The goal is to use cluster to blur an image.

First we'll make use of scipy to read the image in.  The image is translated in the a 3 dimensional array, the x and y describe the hight and width, then the third dimension represents the RGB values for each image.

```python
!wget http://blog.trenthauck.com/assets/headshot.jpg

from scipy import ndimage

img = ndimage.imread("headshot.jpg")

from matplotlib import pyplot as plt
from matplotlib import cm
plt.imshow(img)
```

```python
img.shape
```

```python
x, y, z = img.shape
```

```python
long_img = img.reshape(x*y, z)
```

```python
long_img.shape
```

```python
from sklearn import cluster
```

```python
cluster.KMeans?
```

```python
k_means = cluster.KMeans(n_clusters=5)
```

```python
k_means.fit(long_img)
```

```python
centers = k_means.cluster_centers_
centers
```

```python
labels = k_means.labels_
labels[:5]
```

Once we have the new labels. We expand the centers to be the correct shape.

```python
plt.imshow(centers[labels].reshape(x, y, z))
plt.savefig("978-1-78398-948-5_6.png")
```
