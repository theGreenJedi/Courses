#Getting Ready
In this recipe we'll show how a decomposition method can actually be used for classification.  DictionaryLearning attempts to take a dataset and transform it into a sparse representation.  The idea being that the features are basis for the resulting datasets.  In an effort to keep this recipe short I'll assume you have a idis_data and iris_target ready to go.

#How to do it
First import Dictionary learning.

```python
from sklearn.decomposition import DictionaryLearning
```

Next, we'll use three components to represent the three species of iris.

```python
dl = DictionaryLearning(3)
transformed = dl.fit_transform(iris_data[::2])
transformed[:5]
```
